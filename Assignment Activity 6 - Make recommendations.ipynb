```python
# Load the dataset
tweets = pd.read_csv('tweets.csv')

print(tweets.shape)
print(tweets.dtypes)
tweets.head()
```

```python
healthcare_tweet_read = tweets[tweets['tweet_entities_hashtags'] == '#Healthcare']

healthcare_tweet_read
```

```python
def contains_nhs(x):
    """does the tweet contain nhs?"""
    y = x.lower()
    return "nhs" in y
    
print(contains_nhs(x="nhs"))
print(contains_nhs(x="NHS"))
print(contains_nhs(x="water"))
```

```python
nhs_tweets = tweets['tweet_full_text'].apply(contains_nhs)

print(nhs_tweets)
tweets[nhs_tweets]
```

```python
def contains_uk(x):
    """does the product contain uk?"""
    y = x.lower()
    return "uk" in y
    
print(contains_uk(x="uk"))
print(contains_uk(x="UK"))
print(contains_uk(x="water"))
```

```python
uk_tweets = tweets['tweet_full_text'].apply(contains_uk)

print(uk_tweets)
tweets[uk_tweets]
```

```python
retweet_count = tweets['tweet_retweet_count'].value_counts()

retweet_count = pd.DataFrame(retweet_count)
retweet_count = tweet_count.reset_index()
retweet_count.columns = ['Count of retweets']
```

```python
tweets['tweet_favorite_count'].value_counts()
```

```python
# Create a new DataFrame containing only the text.
tweets_text = tweets[['tweet_full_text']]

# View the DataFrame.
tweets_text.head(5)
```

```python
# # Loop through the messages, and create a list of values containing the # symbol.
tags = []

for y in [x.split(' ') for x in tweets['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z.lower())
```

```python
print(tags)
```

```python
tags_series = pd.Series(tags)

tags_series
```

```python
tag_count = tags_series.value_counts()
```

```python
tweet_count = pd.DataFrame(tag_count)
tweet_count = tweet_count.reset_index()
tweet_count.columns = ['Hashtags', 'Count of Hashtags']

print(tweet_count.shape)
print(tweet_count.dtypes)
tweet_count.head(5)
```

```python
nhs_tweets = tweet_count[tweet_count['Hashtags'] == '#nhs']

nhs_tweets
```

```python
# Slice the DataFrame to any tweets with count > 10
popular_tweets = tweet_count.loc[tweet_count['Count of Hashtags'] > 10]

print(popular_tweets.shape)
popular_tweets.tail(5)
```

```python
# Create a Seaborn barplot indicating records with a count >10 records.
tweet_plot = sns.barplot(x='Hashtags', y='Count of Hashtags', estimator=sum, data=popular_tweets)
plt.title("Most Popular Tweets", y=1.02, fontsize=16);
plt.xticks(rotation=90)
```

```python
# Filter the data set to only look at data from 2021-08 onwards.
ar_filtered = appointments_regional[appointments_regional['appointment_month'] >='2021-08']

ar_filtered.head()
```

```python
# Create an aggregated data set to review the different features.
ar_agg = ar_filtered.groupby(['appointment_month'])[['count_of_appointments']]\
.sum().reset_index().copy()

# View the DataFrame.
ar_agg
```

```python
# Determine the total number of appointments per month.
# Add a new column to indicate the average utilisation of services.
# Monthly aggregate / 30 to get to a daily value.
ar_agg['utilisation'] = ar_agg['count_of_appointments'] / 30
ar_agg['utilisation'] = ar_agg['utilisation'].round(decimals = 1)

# View the DataFrame.
ar_agg
```

```python
# Plot sum of count of monthly visits.
# Create a lineplot with Seaborn.

plt.figure(figsize = (18, 6))
ar_agg = sns.lineplot(x='appointment_month', y='count_of_appointments', data=ar_agg, ci=None)
plt.title("Monthly Visits")
```

```python
# Plot monthly capacity utilisation.
# Create a lineplot.

plt.figure(figsize = (18, 6))
ar_agg = sns.lineplot(x='appointment_month', y='utilisation', data=ar_agg, ci=None)
plt.title("Average Daily Utilisation by Month")
```
