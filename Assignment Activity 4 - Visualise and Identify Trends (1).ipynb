```python
# Create Create four visualisations indicating the number of appointments for service setting per season. 
# The seasons are summer (August 2021), autumn (October 2021), winter (January 2022), and spring (April 2022).
# Create a separate data set that can be used in future weeks. 
# Summer August 2021

nc_summer = national_categories[(national_categories['appointment_date'] >'2021-08-01') & \
                                (national_categories['appointment_date'] <'2021-08-31')]

print(nc_summer.shape)
nc_summer.head()
```

```python
nc_summer_appts = nc_summer.groupby(['service_setting', 'appointment_date'])\
[['count_of_appointments']].sum().reset_index().copy()

nc_summer_appts.head()
```

```python
# Create a lineplot of the data 
# Plot the data
plt.figure(figsize = (18, 6))

nc_plot = sns.lineplot(x='appointment_date', y='count_of_appointments', hue='service_setting',
                       data=nc_summer_appts, ci=None)
plt.title("August 2021 appointment count by service setting")
nc_plot.set_xticklabels(['3', '11', '15', '19', '23', '27'], minor=True)
```
