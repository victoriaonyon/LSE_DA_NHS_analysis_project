```python
# Created a DataFrame to store the 'National Categories' excel file
# Checkes dataset has loaded correctly using shape, type and head() functions
# Used the print functions with check function as this will provide an output for each function
# Used head() function as this gives a snapshot of what the data looks like

national_categories = pd.read_excel('national_categories.xlsx')

print(national_categories.shape)
print(national_categories.dtypes)
national_categories.head()
```

```python
# Determine the descriptive statistics for the National Categories dataset
national_categories.describe()
```

```python
# Determine the metadata for the National Categories dataset
national_categories.info()
```

```python
# Q1. How many locations are there in the data set?
# Data is sorted in desc order using value_counts

national_categories['sub_icb_location_name'].value_counts()
```

```python
# Print the number of locations
print(f"The count of locations is {national_categories['sub_icb_location_name'].count()}")
```

```python
# Q2. What are the five locations with the highest number of records?
# The most popular location is found using idxmax()
national_categories['sub_icb_location_name'].value_counts().idxmax()
```

```python
#Create a DataFrame of the locations 
national_categories['sub_icb_location_name'].value_counts().to_frame('locations')
```

```python
# Print the top 5 locations - as series is already sorted in Desc order
national_categories['sub_icb_location_name'].value_counts().iloc[0:5]
```
