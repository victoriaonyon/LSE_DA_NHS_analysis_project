```python
# Create a new DataFrame containing only the text.
tweets_text = tweets[['tweet_full_text']]

# View the DataFrame.
tweets_text.head(5)
```

```python
# Loop through the messages, and create a list of values containing the # symbol.
tags = []

for y in [x.split(' ') for x in tweets['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z.lower())
```

```python
# Display the first 30 records.
print(tags[1:30])
```

```python
# Convert list to a series 
tags_series = pd.Series(tags)

tags_series
```

```python
# Use value_counts() to get a count of how often each tag appears
tag_count = tags_series.value_counts()
```

```python
# Convert the series to a DataFrame in preparation for visualisation.
tweet_count = pd.DataFrame(tag_count)
tweet_count = tweet_count.reset_index()
tweet_count.columns = ['Hashtags', 'Count of Hashtags']

print(tweet_count.shape)
print(tweet_count.dtypes)
tweet_count.head(5)
```

```python
# Display records where the count is larger than 10
# Use tail() to check all tweets have at more than 10 records
popular_tweets = tweet_count.loc[tweet_count['Count of Hashtags'] > 10]

print(popular_tweets.shape)
popular_tweets.tail(5)
```

```python
# Check the top 5 tweets
popular_tweets.head(5)
```

```python
# Create a Seaborn barplot indicating records with a count >10 records.
# Colorblind palette used to improve readability
sns.set(rc={'figure.figsize':(15, 12)})

tweet_plot = sns.barplot(x="Hashtags", y='Count of Hashtags', estimator=sum, data=popular_tweets,
                        palette='colorblind')
plt.title("Most Popular Tweets", y=1.02, fontsize=16);
plt.xticks(rotation=90)
plt.xlabel("Popular Hashtags", fontsize= 12)
plt.ylabel("Count of Hashtags", fontsize= 12)
```

```python
# Checking for outliers - Select hashtag count between 10 < 41 - removing two extreme outliers
# Create a boxplot to view skew of data
popular_tweets_filtered = popular_tweets[(popular_tweets['Count of Hashtags'] > 10)\
                                         & (popular_tweets['Count of Hashtags'] <= 41)]

fig = plt.figure(figsize=(18, 4))

sns.boxplot(data=popular_tweets_filtered, y='Count of Hashtags',  whis=1.5)
```

The boxplot shows even removing the two extreme outliers that the majority of the hashtags have a median of 15.

```python
# Define a function to map the values
def set_timevalue(row_number, assigned_value):
    return assigned_value[row_number]

lead_dictionary ={'1 Day' : 2, 'Same Day' : 1, '2 to 7 Days' : 7, '22  to 28 Days' : 28, 
                  '8  to 14 Days' : 14, '15  to 21 Days' : 21,'More than 28 Days' : 31, 'Unknown/Data Quality': 100}

lead_ar_filtered['lead_time'] = lead_ar_filtered['time_between_book_and_appointment'].\
apply(set_timevalue, args =(lead_dictionary,))
 
# Print the DataFrame
print(lead_ar_filtered.dtypes)
lead_ar_filtered.head()
```
